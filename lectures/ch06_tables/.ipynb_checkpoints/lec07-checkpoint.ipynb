```python
from datascience import *
import numpy as np

%matplotlib inline
import matplotlib.pyplot as plots
plots.style.use('fivethirtyeight')
```

## Lecture 7 ##

## Census ##

```python
full = Table.read_table('nc-est2019-agesex-res.csv')
```

```python
# Keep only the columns we care about
partial = full.select('SEX', 'AGE', 'POPESTIMATE2014', 'POPESTIMATE2019')
```

```python
### Make things easier to read by relabeling column names (using function "relabeled")

simple = partial.relabeled(2, '2014').relabeled(3, '2019')
simple
```

## Line Plots ##

```python
# Remove the age totals (999)

no_999 = simple.where('AGE', are.below(999))
```

```python
### Our first chart!

overall = no_999.where('SEX', 0)   ### 0 = male and female
# overall
overall.plot('AGE', '2019')

### le6 means 10 to the power of 6 (10^6), which is equal to 1,000,000. 
```

```python
### label/title
### That plot should be labeled! Here are 3 ways to label it:
```

```python
# US Population  <--- Just add a comment

overall.plot('AGE', '2019')
```

```python
overall.plot('AGE', '2019')
print('US Population')  # <--- Print out what it is; can move to above overall plot
```

```python
### print a nice title using "title()"

overall.plot('AGE', '2019')
plots.title('US Population');   ### attribute add to the plot 
```

## Males vs Females

```python
### instead of the whole populatin, we are interested in comparing male and female counts per age
### we are creating variables!!
### where SEX 1 vs 2
### drop 

males = no_999.where('SEX', 1).drop('SEX')
females = no_999.where('SEX', 2).drop('SEX')
# males    ### just checking
```

```python
### note the use of "column()" gives us arrays to be added to the new table
### adding the 2019 arrays to the new table

pop_2019 = Table().with_columns(
    'Age', males.column('AGE'),
    'Males', males.column('2019'),
    'Females', females.column('2019')
)
pop_2019
```

```python
### the plot syntax: tbl.plot(x_column, y_column)
### but the datascience libarary makes it easy and we only give "Age"

pop_2019.plot('Age')
```

```python
# Calculate the percent female for each age

total = pop_2019.column('Males') + pop_2019.column('Females')
pct_female = pop_2019.column('Females') / total * 100
pct_female

### we have an array because you take out one "column"
```

```python
# Round it to 3 so that it's easier to read
pct_female = np.round(pct_female, 3)
pct_female
```

```python
# Add female percent to our table
pop_2019 = pop_2019.with_column('Percent female', pct_female)
pop_2019
```

```python
### the standard plot() syntax

pop_2019.plot('Age', 'Percent female')

### what does this line plot mean???
```

```python
# ^^ Look at the y-axis! Trend is not as dramatic as you might think!!

pop_2019.plot('Age', 'Percent female')
plots.ylim(0, 100);  # the y-limits of current axes

### ylim: see https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.ylim.html
```

## Bar Charts ##

```python
# Highest grossing movies as of 2017
top_movies = Table.read_table('top_movies_2017.csv')
top_movies
```

```python
top10_adjusted = top_movies.take(np.arange(10))
top10_adjusted
```

```python
# Convert to millions of dollars for readability
millions = np.round(top10_adjusted.column('Gross (Adjusted)') / 1000000, 3)
top10_adjusted = top10_adjusted.with_column('Millions', millions)
top10_adjusted
```

```python
### A line plot/chart/graph doesn't make sense here: don't do this!

top10_adjusted.plot('Year', 'Millions')
```

```python
### choose the right variables
### using the barh() method

top10_adjusted.barh('Title', 'Millions')
```
