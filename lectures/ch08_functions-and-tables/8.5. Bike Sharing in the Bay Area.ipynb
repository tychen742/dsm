# 8.5. Bike Sharing in the Bay Area

- Bike rental company
- September 2014 - August 2015
- 354,152 rentals

```python
##### header
path_data = '../../data/'
from datascience import Table
from datascience.predicates import are
##### Predicate functions are functions that return a single TRUE or FALSE.
%matplotlib inline
```

```python
trips = Table.read_table(path_data + 'trip.csv')
trips.show(5)
```

```python
##### free trips: trips that last less than 1800 seconds (half an hour)
##### there will be more interesting stuff if we do an analysis on all data, starting with descriptive. 
commute = trips.where('Duration', are.below(1800))
commute.hist('Duration', unit='Second')
```

```python
commute.hist('Duration')
```

```python
##### use more bins to get more details
commute.hist('Duration', bins=60, unit='Second')
```

## 8.5.1. Exploring the Data with group and pivot

Use group() to identify the most highly used Start Station:

```python
commute.group(['Start Station', 'End Station'])
```

## 8.5.2. Drawing Maps

With the pivot table, we can map the data because we have the lat & long of the stations

```python
stations = Table.read_table(path_data + 'station.csv')
stations
```

```python
##### Marker.map_table: We don't usually write our own functions
##### https://www.data8.org/datascience/maps.html
### from datascience import *     ==> will confuse itself
# from datascience.maps import Marker
from datascience.maps import *
Marker.map_table(stations.select('lat', 'long', 'name').relabel('name', 'labels'))
# Marker.map_table(stations.select('lat', 'long', 'name'))
```

```python
sf = stations.where('landmark', are.equal_to('San Francisco'))
sf_map_data = sf.select('lat', 'long', 'name').relabel('name', 'labels')
Circle.map_table(sf_map_data, color='green')
```

## 8.5.3. More Informative Maps: An Application of join

```python
##### aggregatoin: GROUP BY 
cities = stations.group('landmark').relabel('landmark', 'city')
cities
```

```python
from datascience import *

colors = cities.with_column('color', make_array('blue','red','green','orange','purple'))
colors
```

```python
##### join stations and colors by landmark
### https://www.data8.org/datascience/_autosummary/datascience.tables.Table.relabel.html#datascience.tables.Table.relabel

joined = stations.join('landmark', colors, 'city')
colored = joined.select('lat', 'long', 'name', 'color').relabel('name', 'labels')
Marker.map_table(colored)
```

```python
##### Where most of the bike rentals originate?
### group is expensive
starts = commute.group('Start Station').sort('count', descending=True)
starts
```

```python
##### geographical data 
station_starts = stations.join('name', starts, 'Start Station')
station_starts
```

```python
##### extract data needed for drawing map, 
##### adding a color and an area to each statio

starts_map_data = station_starts.select('lat', 'long', 'name').with_columns(
    'colors', 'blue',
    'area', station_starts.column('count') * 0.3
)
starts_map_data.show(3)
Circle.map_table(starts_map_data.relabel('name', 'labels'))
```
