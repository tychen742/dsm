```python
##### but doesn't this value on the very left skew the distribution?
##### we need to remove it and re-graph to see.
no_outlier = actors.where('Number of Movies', are.above(10))
no_outlier.scatter('Number of Movies', 'Average per Movie', fit_line=True)

##### the negative association is true
```

```python
##### now, we want to see the top money-making actors:

actors.where('Number of Movies', are.above(60)).sort('Gross', descending=True)
```
