## Scatter Plots
A variable is a formal name for what we have been calling a “feature” or “attribute”, such as ‘number of movies.’ 

```python
### The Table method scatter draws a scatter plot consisting of one point for each row of the table.
### tabulating 2 variables: same person/dot x (col1, col2)
actors.scatter('Number of Movies', 'Total Gross')   ### requires %plot   ### run 3 times

### analysis (guess): positive association? 
```

```python
### Number of Movies' X 'Average per Movie
actors.scatter('Number of Movies', 'Average per Movie')

########## negative association!!!!!
########## outlier
```

```python
##### For the right end, let’s zoom in on the main body of the plot 
##### by just looking at the portion that doesn’t have the outlier.
no_outlier = actors.where('Number of Movies', are.above(10))
no_outlier.scatter('Number of Movies', 'Average per Movie')

##### slice num of movies > 10; no outliers
##### The negative association is still clearly visible
```

```python
##### who are in more movies?
actors.where('Number of Movies', are.above(60))

##### Robert DeNiro has the highest number of movies and the lowest average receipt per movie.
##### an extreme case
```

```python
#####  Let us now take a look at the outlier.
actors.where('Number of Movies', are.below(10))

##### 7 movies; 452 million dollars per movie.
```

# Line Plots

```python
from IPython.display import Image
Image(filename="movies_by_year.jpg")
```

```python
movies_by_year = Table.read_table('movies_by_year.csv')
movies_by_year
```

```python
##### The Table method plot produces a line plot.
movies_by_year.plot('Year', 'Number of Movies')
```

```python
century_21 = movies_by_year.where('Year', are.above(1999))
century_21.plot('Year', 'Number of Movies')

##### The global financial crisis of 2008 has a visible effect – in 2009 there is a sharp drop in the number of movies released.
```

```python
century_21.plot('Year', 'Total Gross')
```

```python
##### 2009 movies
century_21.where('Year', are.equal_to(2009))
```
